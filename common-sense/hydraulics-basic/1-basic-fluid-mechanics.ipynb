# Dimensions, Terminology, and Review of Basic Fluid Mechanics

明渠有开放界面，一般是水和空气交接，且面积和深度有关，求解过程需要处理几何问题，比管道流体更复杂。

首先了解不同维度flow的基本概念。

## 1-，2-，3- 维 flows

flow 的维度是描述流体所需的互不相关的空间变量的数量。如果只需要一个方向x（定义为沿渠道的坐标轴坐标）就能描述流体，那么流体就是1维的，这样的话，depth Y 和velocity V 都可以定义为 Y(x), V(x)，二维就是x和另外一个描述从渠道底部向上的坐标轴坐标y或者横向跨过渠道的坐标轴坐标z，三维就是x，y，z都有。 如果向量随时间变化，那么就再多一个t。二维随时间变化的时候虽然实际上是三维的，但在水力学里面，还是称之为二维的，只不过是unsteady，即非平稳的。实际二维的就称为平稳的。

所以最简单的问题自然就是一维平稳流了。术语hydraulics of open channel flow 通常用来表示一维free surface flow，也就是说，可以定义明渠流量为过流断面平均流速，且不用考虑断面上流速随深度和位置的变化，V(x)=Q/A，即使有时候为了更精确表示动能还会引入一些修正系数（稍后补充），假设仍然可以维持。

## Steady 与 Unsteady flow

时不变的flow就是平稳的。数学上来说，即$\partial Y/\partial t=0$,$\partial V/\partial t=0$,$\partial A/\partial t=0$，即depth of flow，the velocity，以及断面面积都是不随时间变化的。

当任意变量是时变的时候，就是unsteady flow了。水力学里面应该更多用定常/非定常来翻译。

## Uniform 与 Nonuniform flow

如果任意变量都随位置不变，就是均匀流，即$\partial Y/\partial x=0$,$\partial Y/\partial x=0$,$\partial Y/\partial x=0$；类似地，任意变量随位置变化，就是nonuniform的了。

nonuniform flows可以进一步分为 gradually varied（渐变流），rapid varied（急变流），和spatially varied。当河流很直的时候，nonuniform flow 可以被当作 gradually varied，渐变流中水压会随深度增加，这种变化是有规律的，称为hydrostatic。rapidly flow压力的分布则是not hydrostatic的。

渐变流是可以用一维水动力方程表示的，rapidly flow则必须用二维或者三维的表示。由于二维三维难解，所以一般会用一维加实验系数修正的方式来求解rapidly flow。

不过并没有明确的定义来区分渐变流和急变流，是有些主观的。不同几乎所有人都同意流经一个大坝溢洪道的肯定是急变流，而上游的水流则是渐变的。过闸门的水流也急变，不过流过之后慢慢又会变为渐变。

spatially varied 流体是主干流和支流有交互的地方发生。因此，flow rate 会随着channel位置变化。也就是说，在两个channels交汇处会有一个短距离上的空间变化。

## Prismatic 与 Nonprismatic channels

和uniform和nonuniform定义类似，不过是针对channel而言的。随空间位置断面几何形状不变的channel就是prismatic的。

## Subcritical，Critical 和 Supercritical flows

首先需要了解一个概念，重力波 gravity wave，是在液体介质内或两种介质界面间（例如大气与海洋间）的一种波，其恢复力来自于重力或浮力。当一小团液体离开液面（界面类型）或者在液体中到了一个液体密度不同之区域（液体内类型），透过重力作用，这团液体会以波动形式在平衡态之间摆荡。比如风浪。即收到干扰离开了原位置，再由于重力重新回到原位置，但是有惯性会波动，形成的波。一个小幅度重力波的波速可以由下式给出：
$$c=\sqrt{\frac{gA}{T}}=\sqrt{gY_d}$$
其中，g是重力加速度，A是过流断面，T是流体顶层宽度，Yd称为水力深度

subcritical flow就是流体会注意到下游变化的影响，并根据下游发生的预期进行调整。也就是说如果有一个障碍物，那么流体深度会逐渐增加到能漫过这个障碍物。这种重力波能够逆流而上，因为其速度比流速小。

而supercritical flow则不同，变化的影响不能向上游传播，因为水道中的速度超过了重力波的传播速度。因此，流量不会根据下游的情况自行调整。例如，如果一个含有超临界流体的水道突然结束，那么在自由溢流处的水道末端的深度将与该水道继续流过时的深度相同。相反，当水流为亚临界时，水流深度在自由溢流结束时向临界深度方向减小。

根据重力波能否向上游移动，亚临界流可以控制下游，而超临界流则可以控制上游。下游和上游控制的例子都存在于通道中的闸门处。从闸门逆流而上的水流必须是亚临界的，因为位于下游的闸门控制水流的深度、流速、面积等。从闸门开口向下的水流将是超临界的，开口决定了水流变量的大小。例如，如果闸门降低，在增加下游流速的同时，会降低下游的深度，对上游的流量产生相反的影响。

Froude number 可以反映流速和波速的相对大小：
$$F_r=\frac{V}{c}=\frac{V}{\sqrt{gA/T}}=\sqrt{\frac{Q^2T}{gA^3}}$$
Froude number为1的flow就是critical flow。

## Turnulent 和 Laminar flow

雷诺数是探讨湍流和层流区别的值。
$$R_e=\frac{VR_n}{v}=\frac{Q}{vP}=\frac{\rho VR_h}{\mu}$$
其中，

V是平均流速，Q是流量，$\mu$是绝对黏度，v是流体的kinematic 黏度。黏度更像是流体内的摩擦力，理想流体或无黏度流体是没有阻力来对抗剪应力的流体。Rh是水力半径即断面面积与湿周之比。

雷诺数小于500时候，流体是层流。
